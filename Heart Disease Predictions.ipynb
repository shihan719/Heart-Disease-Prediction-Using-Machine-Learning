<span style="font-family:Calibri">Data Processing</span>

<span style="font-family:Calibri">After exploring the dataset, I noticed the necessity to convert certain categorical variables into dummy variables and scale all values before training the machine learning models. Initially, I will utilize the "get_dummies" method to generate dummy columns for categorical variables.</span>

```python
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
y = dataset['target']
X = dataset.drop(['target'], axis = 1)
```

<span style="font-family:Calibri"><span style="color:yellow">K Neighbors Classifier</span>

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 12)
score=cross_val_score(knn_classifier,X,y,cv=10)
score.mean()
```

<span style="font-family:Calibri"><span style="color:orange">Random Forest Classifier</span>

```python
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier= RandomForestClassifier(n_estimators=10)
score2=cross_val_score(randomforest_classifier,X,y,cv=10)
score2.mean()
```

<span style="font-family:Calibri"><span style="color:orange">Decision Tree Classifier</span>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
decisiontree_classifier = DecisionTreeClassifier()
score = cross_val_score(decisiontree_classifier, X, y, cv=10)
score.mean()
```

<span style="font-family:Calibri"><span style="color:orange">XGB Classifier</span>

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

# Assuming X and y are your features and target variable, respectively
xgb_classifier = XGBClassifier()
score = cross_val_score(xgb_classifier, X, y, cv=10)
score.mean()

# mean_score = score.mean()

# print("Mean Cross-Validation Score:", mean_score)
```

<span style="font-family:Calibri">Bar Plot</span>

```python
# K-Nearest Neighbors
knn_classifier = KNeighborsClassifier(n_neighbors=12)
score_knn = cross_val_score(knn_classifier, X, y, cv=10).mean()

# Random Forest
randomforest_classifier = RandomForestClassifier(n_estimators=10)
score_rf = cross_val_score(randomforest_classifier, X, y, cv=10).mean()

# Decision Tree
decisiontree_classifier = DecisionTreeClassifier()
score_dt = cross_val_score(decisiontree_classifier, X, y, cv=10).mean()

# XGBoost
xgb_classifier = XGBClassifier()
score_xgb = cross_val_score(xgb_classifier, X, y, cv=10).mean()

# Plotting
classifiers = ['KNN', 'Random Forest', 'Decision Tree', 'XGBoost']
scores = [score_knn, score_rf, score_dt, score_xgb]

fig, ax = plt.subplots()
bars = ax.bar(classifiers, scores, color=['SteelBlue', 'ForestGreen', 'IndianRed', 'MediumPurple'])

# Annotate bars with their sum values
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.01, round(yval, 4), ha='center', va='bottom')

# # Highlight the highest value
# max_index = scores.index(max(scores))
# bars[max_index].set_color('gold')


# Finding the index of the highest score
max_score_index = scores.index(max(scores))
# Getting the classifier .201


plt.xlabel('Classifiers')
plt.ylabel('Cross-Validation Score')
plt.title('Comparison of Classifier Performance with Sum Values')
plt.show()
```

<span style="font-family:Calibri">The K-Nearest Neighbors (KNN) model outperformed other classifiers, achieving the highest cross-validation score. This suggests that KNN, with 12 neighbors, is well-suited for the given classification task based on the provided dataset.</span>
